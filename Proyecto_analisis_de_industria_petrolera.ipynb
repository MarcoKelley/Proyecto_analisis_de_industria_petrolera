Entrenamiento del modelo


```python

target = geo_data_0['product']
features = geo_data_0.drop(['product','id'], axis=1)
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345
)
```

Predicciones para el conjunto de validacion 

```python

model = LinearRegression()
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)
```

Almacenamiento de las predicciones y las respuestas correctas para el conjunto de validacion


```python

correct_answers = target_valid
predictions = predicted_valid
```

Mostrando el volumen medio de reservas predicho y RMSE del modelo 

```python
print(predictions.mean())
```

```python
rmse = np.sqrt(mean_squared_error(target_valid, predicted_valid))
print(rmse)
```

Ganancias por pozos 

Region 0 

```python
def calcular_ganancia(predicciones_200_pozos):
    ingresos_por_pozo = predicciones_200_pozos * 4500  # 4500 USD por unidad
    costo_por_pozo = 100_000_000 / 200  # 500,000 USD por pozo
    ganancia_por_pozo = ingresos_por_pozo - costo_por_pozo
    ganancia_total = ganancia_por_pozo.sum()
    return ganancia_total
```

```python
ganancia_region_0 = calcular_ganancia(predictions[mejores_pozos_0])
print(ganancia_region_0)
```

Region 1

```python
ganancia_region_1 = calcular_ganancia(predictions_1[mejores_pozos_1])
print(ganancia_region_1)
```

Region 2

Calculo de riesgos y ganancias por region 

Region 0

```python
valores_mejores_pozos_0 = correct_answers.iloc[mejores_pozos_0]
valores_mejores_pozos_1 = correct_answers.iloc[mejores_pozos_1]
valores_mejores_pozos_2 = correct_answers.iloc[mejores_pozos_2]
```

```python
def bootstrap_region_0(n_samples=1000):
    ganancias = []
    for i in range(n_samples):
        indices_aleatorios = np.random.choice(len(valores_mejores_pozos_0), size=200, replace = True)
        ganancia = calcular_ganancia(valores_mejores_pozos_0.iloc[indices_aleatorios])
        ganancias.append(ganancia)
    return np.array(ganancias)

    
```

```python

ganancias_bootstrap_0 = bootstrap_region_0()
print(f"Número de simulaciones: {len(ganancias_bootstrap_0)}")
print(f"Ganancia promedio: ${ganancias_bootstrap_0.mean()/1_000_000:.2f}M")

```

```python
print(f"Ganancia mínima: ${ganancias_bootstrap_0.min()/1_000_000:.2f}M")
print(f"Ganancia máxima: ${ganancias_bootstrap_0.max()/1_000_000:.2f}M")
print(f"Percentil 2.5%: ${np.percentile(ganancias_bootstrap_0, 2.5)/1_000_000:.2f}M")
print(f"Percentil 97.5%: ${np.percentile(ganancias_bootstrap_0, 97.5)/1_000_000:.2f}M")

# Probabilidad de pérdidas
prob_perdidas = (ganancias_bootstrap_0 < 0).mean() * 100
print(f"Probabilidad de pérdidas: {prob_perdidas:.1f}%")
```
